````python
# Chart - 6 visualization code

def compare_default_across_features(df, feature_column, feature_labels):
    # Group the data by the specified feature column and calculate the count of defaults and non-defaults for each feature level
    feature_defaults = df.groupby([feature_column, 'default']).size().unstack()

    # Setting the figure size
    plt.figure(figsize=(10, 5))

    # Plotting the stacked bar chart to compare default status across the specified feature levels
    feature_defaults.plot(kind='bar', stacked=True, color=['lightblue', 'lightcoral'], edgecolor='k')

    # Setting the custom x-axis labels based on the provided mapping (feature_labels)
    x_labels = [feature_labels[label] for label in feature_defaults.index]
    plt.xticks(ticks=range(len(x_labels)), labels=x_labels, rotation=45, ha='right')

    # Setting the y-axis label
    plt.ylabel('Count', fontsize=14)

    # Setting the title of the plot
    plt.title('Credit Card Default Status across ' + feature_column.capitalize(), fontsize=16)

    # Displaying the plot
    plt.show()

# Mapping of education levels to custom labels
education_labels = {
    0: "0",
    1: "Grad School",
    2: "University",
    3: "High School",
    4: "Others",
    5: "5",
    6: "6",
}

# Call the function to compare default status across education levels using the custom labels
compare_default_across_features(df, "education", education_labels)
````
