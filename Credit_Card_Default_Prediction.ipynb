````python
# Chart - 6 visualization code

def compare_default_across_features(df, feature_column, feature_labels):
    # Group the data by the specified feature column and calculate the count of defaults and non-defaults for each feature level
    feature_defaults = df.groupby([feature_column, 'default']).size().unstack()

    # Setting the figure size
    plt.figure(figsize=(10, 5))

    # Plotting the stacked bar chart to compare default status across the specified feature levels
    feature_defaults.plot(kind='bar', stacked=True, color=['lightblue', 'lightcoral'], edgecolor='k')

    # Setting the custom x-axis labels based on the provided mapping (feature_labels)
    x_labels = [feature_labels[label] for label in feature_defaults.index]
    plt.xticks(ticks=range(len(x_labels)), labels=x_labels, rotation=45, ha='right')

    # Setting the x-axis label
    plt.xlabel('Education', fontsize=14,color='blue')

    # Setting the y-axis label
    plt.ylabel('Count', fontsize=14,color='blue')

    # Setting the title of the plot
    plt.title('Credit Card Default Status across ' + feature_column.capitalize(), fontsize=16,color='red')

    # Displaying the plot
    plt.show()

# Mapping of education levels to custom labels
education_labels = {
    0: "0",
    1: "Grad School",
    2: "University",
    3: "High School",
    4: "Others",
    5: "5",
    6: "6",
}

# Call the function to compare default status across education levels using the custom labels
compare_default_across_features(df, "education", education_labels)
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

**Hypothesis statement 1:** There is a significant difference in the credit limit between customers who default on their credit card payments and those who do not.

**Hypothesis statement 2:** The age distribution of customers who default on their credit card payments is different from the age distribution of customers who do not default.

**Hypothesis statement 3:** There is a correlation between the amount of the last bill statement ('bill_amt1') and the amount paid in the last payment ('pay_amt1').

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Research Hypothesis:** The credit limit is different between customers who default on their credit card payments and those who do not.

**Null Hypothesis (H0):** There is no significant difference in the credit limit between default and non-default cases.

**Alternate Hypothesis (Ha):** There is a significant difference in the credit limit between default and non-default cases.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Separate credit limits for default and non-default cases
default_limit = df[df['default'] == 1]['limit_bal']
non_default_limit = df[df['default'] == 0]['limit_bal']

# Perform the two-sample independent t-test
t_statistic, p_value = stats.ttest_ind(default_limit, non_default_limit, equal_var=False)

# Display the p-value
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference in the credit limit between default and non-default cases.")
else:
    print("There is no significant difference in the credit limit between default and non-default cases.")
````

##### Which statistical test have you done to obtain P-Value?

To obtain the p-value, I performed a two-sample independent t-test. The t-test is used to compare the means of two independent groups to determine if there is a statistically significant difference between them. In this case, I used the t-test to compare the credit limits of two groups: customers who defaulted on their credit card payments and those who did not. The p-value from the t-test helps us assess whether the observed difference in credit limits between the two groups is statistically significant or just due to random chance. If the p-value is below a chosen significance level (e.g., 0.05), we can reject the null hypothesis and conclude that there is a significant difference in credit limits between default and non-default cases.

##### Why did you choose the specific statistical test?

I chose the specific statistical test, the two-sample independent t-test, because it is suitable for comparing the means of two independent groups, which is the case for comparing credit limits between customers who defaulted and those who did not. The t-test is commonly used for continuous variables with binary categorical variables in independent groups.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Research Hypothesis:** The age distribution of customers who default on their credit card payments is different from the age distribution of customers who do not default.

**Null Hypothesis (H0):** There is no significant difference in the age distribution between customers who default and those who do not default.

**Alternate Hypothesis (Ha):** There is a significant difference in the age distribution between customers who default and those who do not default.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Separate age values for default and non-default cases
default_age = df[df['default'] == 1]['age']
non_default_age = df[df['default'] == 0]['age']

# Perform the Mann-Whitney U test
u_statistic, p_value = stats.mannwhitneyu(default_age, non_default_age, alternative='two-sided')

# Display the p-value
print("P-value:", p_value)

if p_value < 0.05:
    print("The age distribution of customers who default is different from non-default cases.")
else:
    print("The age distribution of customers who default is not significantly different from non-default cases.")
````

##### Which statistical test have you done to obtain P-Value?


To obtain the p-value, I performed the Mann-Whitney U test. The Mann-Whitney U test is a non-parametric test used to compare the distributions of two independent groups. In this case, I used the Mann-Whitney U test to compare the age distribution of customers who defaulted on their credit card payments with the age distribution of customers who did not default.

The Mann-Whitney U test does not assume normality of the data and is appropriate when the assumptions of the t-test (parametric test) are not met. It assesses whether the distributions of the two groups are significantly different from each other.

The p-value obtained from the Mann-Whitney U test indicates the probability of observing the observed difference in age distributions (or an even more extreme difference) if the null hypothesis (no significant difference in age distributions) were true. If the p-value is less than the chosen significance level (e.g., 0.05), we can reject the null hypothesis and conclude that there is a significant difference in the age distribution between customers who default and those who do not.

##### Why did you choose the specific statistical test?

I chose the specific statistical test, the Mann-Whitney U test, because it is appropriate for comparing the distributions of two independent groups when the data may not be normally distributed. Since we are comparing the age distribution of customers who defaulted and those who did not, and the age data might not follow a normal distribution, the Mann-Whitney U test is a suitable non-parametric test to assess if there is a significant difference between the two groups.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Research Hypothesis:** There is a correlation between the amount of the last bill statement ('bill_amt1') and the amount paid in the last payment ('pay_amt1').

**Null Hypothesis (H0):** There is no correlation between 'bill_amt1' and 'pay_amt1'.

**Alternate Hypothesis (Ha):** There is a correlation between 'bill_amt1' and 'pay_amt1'.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Calculate the Pearson correlation coefficient and its associated p-value
correlation, p_value = stats.pearsonr(df['bill_amt1'], df['pay_amt1'])

# Display the correlation coefficient and the p-value
print("Correlation Coefficient:", correlation)
print("P-value:", p_value)

if p_value < 0.05:
    print("There is a significant correlation between 'bill_amt1' and 'pay_amt1'.")
else:
    print("There is no significant correlation between 'bill_amt1' and 'pay_amt1'.")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

I used the Interquartile Range (IQR) method for outlier treatment. The IQR method is a robust approach to identify and handle outliers in a dataset. Here's why I chose this technique and some details about its implementation:

**Outlier Treatment Technique: Interquartile Range (IQR) Method**

**Robustness:** The IQR method is less sensitive to extreme values compared to methods like mean and standard deviation. It uses the middle 50% of the data, making it more robust to the influence of outliers.

**Simple and Easy to Implement:** The IQR method is straightforward to implement and does not require complex calculations. It involves calculating the first quartile (Q1), third quartile (Q3), and the IQR (Q3 - Q1) to define the lower and upper bounds for outlier detection.

**Retains Data Distribution:** The IQR method identifies outliers based on the spread of the data, allowing us to retain the original distribution as much as possible.

**Non-parametric Approach:** The IQR method is non-parametric, meaning it does not assume any specific underlying data distribution. This makes it suitable for datasets with non-normally distributed data.

**Steps of IQR Outlier Treatment:**

1.Calculate the first quartile (Q1) and third quartile (Q3) of the data.

2.Calculate the IQR as the difference between Q3 and Q1 (IQR = Q3 - Q1).

3.Define lower bound as Q1 - 1.5 * IQR and upper bound as Q3 + 1.5 * IQR.

4.Replace any data points that fall below the lower bound or above the upper
bound with the respective bound values.

##### Which hyperparameter optimization technique have you used and why?

 I used the RandomizedSearchCV hyperparameter optimization technique. RandomizedSearchCV is a method that performs a randomized search over specified hyperparameter values for an estimator. It randomly selects a combination of hyperparameters from the specified parameter grid and evaluates the estimator's performance using cross-validation. This approach has several advantages:

**1.Reduced Computation Time:** Compared to GridSearchCV, which exhaustively searches all possible combinations of hyperparameters, RandomizedSearchCV randomly samples a subset of hyperparameter combinations. This reduces computation time, especially when dealing with a large number of hyperparameters and a large dataset.

**2.Efficient Exploration:** RandomizedSearchCV explores a wider range of hyperparameter values, which can be beneficial when we are uncertain about the best hyperparameter values. It allows us to focus on promising regions in the hyperparameter space without wasting computational resources on unpromising combinations.

**3.Better Generalization:** RandomizedSearchCV helps prevent overfitting to the validation set, as it does not exhaustively search for hyperparameter values that optimize the validation set performance. It selects hyperparameters randomly, leading to better generalization to unseen data.

**4.Parallelization:** RandomizedSearchCV can run the evaluations of different hyperparameter combinations in parallel, making it more efficient when working with multi-core processors.

### 3. Explain the model which you have used and the feature importance using any model explainability tool?

The model used for credit card default prediction is the Random Forest classifier. Random Forest is an ensemble learning method that combines multiple decision trees to make predictions. Each decision tree in the forest is trained on a random subset of the data and a random subset of features, and the final prediction is obtained by averaging or voting the predictions of individual trees.

Explainability of the Random Forest model can be achieved by analyzing the feature importances. Feature importances tell us which features (or variables) have the most significant influence on the model's predictions. There are several ways to interpret feature importances in a Random Forest:

**Mean Decrease Impurity (MDI):** This method calculates the feature importances based on how much the impurity (e.g., Gini impurity) is reduced by using each feature in the construction of decision trees.

**Mean Decrease Accuracy (MDA):** MDA evaluates the feature importances by measuring how much the accuracy of the model decreases when a specific feature is permuted or shuffled.

**Permutation Importance:** Similar to MDA, permutation importance evaluates feature importances by shuffling the values of a single feature and measuring the drop in model performance.

**Tree-based Feature Importance:** For each tree in the forest, the feature importance is calculated based on the number of times a feature is selected for splitting and the improvement it brings to the model's performance.
